Seperating feature and target varaiable dropping columns name,

Preprocessing

```python
X = parkinson_data.drop(columns=['name' , 'status'], axis=1)
Y = parkinson_data['status']
```

```python
X
```

```python
Y
```

Applying train_test_split for Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test, Y_train, Y_test = train_test_split( X , Y , test_size=0.2, random_state= 2)
```

```python
X_train.shape
```

```python
Y_test.shape
```

```python
Y_train.shape
```

Now appliying Standardization that will scale the data:



```python
from sklearn.preprocessing import StandardScaler
```

```python
s_scale = StandardScaler()
```

```python
s_scale.fit(X_train)
```

```python
X_train
```

```python
X_train = s_scale.transform(X_train)
```

```python
X_test = s_scale.transform(X_test)
```

Model Traning 

using SVM model

```python
from sklearn import svm
```

```python
model = svm.SVC(kernel= 'linear')
```

Training the model

```python
model.fit(X_train, Y_train)
```

Now Evaluting the model


```python
from sklearn.metrics import accuracy_score
```

```python
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
X_test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('The accuraccy rate is', test_accuracy)
```
